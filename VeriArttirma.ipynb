```python
def save_image(image, folder, filename):
    os.makedirs(folder, exist_ok=True)
    path = os.path.join(folder, filename)
    cv2.imwrite(path, image)
    print(f"Kaydedildi: {path}")
```

```python
def rotate_image(image, angle):
    h, w = image.shape[:2]
    center = (w//2, h//2)
    matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
    rotated = cv2.warpAffine(image, matrix, (w, h))
    return rotated
```

```python
rotated_img = rotate_image(img, 45)
plt.imshow(cv2.cvtColor(rotated_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
save_dir = r"C:\veri arttirimi"
os.makedirs(save_dir, exist_ok=True)
save_path = os.path.join(save_dir, "cat.jpg")
result = cv2.imwrite(save_path, img)
print("OpenCV kaydetme sonucu:", result)
```

```python
def flip_image(image, flip_code):    
    return cv2.flip(image, flip_code)

```

```python
flipped_img = flip_image(img, 1)
plt.imshow(cv2.cvtColor(flipped_img, cv2.COLOR_BGR2RGB))
plt.title("Flip Horizontal")
plt.axis('off')
plt.show()
save_dir = r"C:\temp\veri_arttirma"
os.makedirs(save_dir, exist_ok=True)
save_path = os.path.join(save_dir, "cat.jpg")
result = cv2.imwrite(save_path, img)
print("OpenCV kaydetme sonucu:", result)
```

```python
def add_noise(image, noise_std):
    noise = np.random.normal(0, noise_std, image.shape).astype(np.uint8)
    noisy_image = cv2.add(image, noise)
    return noisy_image
    
```

```python
noisy_img = add_noise(img, 25)
plt.imshow(cv2.cvtColor(noisy_img, cv2.COLOR_BGR2RGB))
plt.title("Noise Std 25")
plt.axis('off')
plt.show()
save_dir = r"C:\temp\veri_arttirma"
os.makedirs(save_dir, exist_ok=True)
save_path = os.path.join(save_dir, "cat.jpg")
result = cv2.imwrite(save_path, img)
print("OpenCV kaydetme sonucu:", result)
```

```python
def adjust_contrast(image, alpha, beta):
    return cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
```

```python
contrast_img = adjust_contrast(img, alpha=1.8, beta=20)
plt.imshow(cv2.cvtColor(contrast_img, cv2.COLOR_BGR2RGB))
plt.title("Contrast α=1.8 β=20")
plt.axis('off')
plt.show()
save_dir = r"C:\temp\veri_arttirma"
os.makedirs(save_dir, exist_ok=True)
save_path = os.path.join(save_dir, "cat.jpg")
result = cv2.imwrite(save_path, img)
print("OpenCV kaydetme sonucu:", result)
```
